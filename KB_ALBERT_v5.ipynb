### train data 불러오기

```python
import pandas as pd

train_df = pd.read_excel('/content/drive/MyDrive/news_train_코사인유사도_전처리 후.xlsx')
test_df = pd.read_excel('/content/drive/MyDrive/news_test_코사인유사도_전처리 후.xlsx')
```

```python
# 텍스트 변환 함수
def replace_symbols(text):
    text = text.replace("↑", "상승")
    text = text.replace("↓", "하락")
    text = text.replace("上", "상승")
    text = text.replace("下", "하락")
    text = text.replace("比", "대비")
    return text

# 텍스트 변환\
train_df['processed_text'] = train_df['texts'].apply(replace_symbols)
test_df['processed_text'] = test_df['texts'].apply(replace_symbols)
```

##### 리스트 형태로 변환하기 -> 토큰화가 List[str] 형태로만 가능

```python
X_train = train_df['processed_text'].tolist()
y_train = train_df['labels'].tolist()

X_test = test_df['processed_text'].tolist()
y_test = test_df['labels'].tolist()
```

### KB-ALBERT 모델 불러오기

```python
from transformers import AutoModel, AutoTokenizer, AutoModelForSequenceClassification

# Load Tokenizer and Model
pt_model = AutoModel.from_pretrained('/content/drive/MyDrive/kb-albert-char-base-v2')
tokenizer = AutoTokenizer.from_pretrained('/content/drive/MyDrive/kb-albert-char-base-v2')
model = AutoModelForSequenceClassification.from_pretrained("/content/drive/MyDrive/kb-albert-char-base-v2", num_labels=2)
```

```python
# 토큰화
train_input = tokenizer(X_train, truncation = True, padding = True,
                        return_tensors="pt")

test_input = tokenizer(X_test, truncation = True, padding = True,
                        return_tensors="pt")
```

```python
import torch

class OurDataset(torch.utils.data.Dataset):
  def __init__(self, inputs, labels):
    self.inputs = inputs
    self.labels = labels

  def __getitem__(self, idx):
    item = {key: torch.tensor(val[idx]) for key, val in self.inputs.items()}
    item['labels'] = torch.tensor(self.labels[idx])
    return item

  def __len__(self):
    return len(self.labels)
```

```python
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }
```

```python
# Dataset 생성
train_dataset = OurDataset(train_input, y_train)
test_dataset = OurDataset(test_input, y_test)

# 데이터로더 생성
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, shuffle = True, batch_size = 8)
test_loader = DataLoader(test_dataset, batch_size = 16)
```

```python
import optuna
from transformers import Trainer
from transformers import TrainingArguments

def objective(trial):
    num_train_epochs = trial.suggest_int('num_train_epochs', 1, 4)
    per_device_train_batch_size = trial.suggest_categorical('per_device_train_batch_size', [8, 16, 32, 64])
    per_device_eval_batch_size = trial.suggest_categorical('per_device_eval_batch_size', [8, 16, 32, 64])
    logging_steps = trial.suggest_categorical('logging_steps', [100, 200, 500])

    training_args = TrainingArguments(
        output_dir='./results',
        num_train_epochs=num_train_epochs,
        per_device_train_batch_size=per_device_train_batch_size,
        per_device_eval_batch_size=per_device_eval_batch_size,
        warmup_steps=500,
        weight_decay=0.01,
        logging_dir='./logs',
        logging_steps=logging_steps
    )

        # Trainer 객체 생성
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=test_dataset,
        compute_metrics=compute_metrics,
    )

    # 모델 훈련 및 평가
    trainer.train()
    eval_result = trainer.evaluate(eval_dataset=test_dataset)
    accuracy = eval_result['eval_accuracy']

    return accuracy

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=10)

# 최적의 하이퍼파라미터 출력
print("Best trial:")
trial = study.best_trial

print("  Value: ", trial.value)
print("  Params: ")
for key, value in trial.params.items():
    print("    {}: {}".format(key, value))
```

```python
from transformers import Trainer

trainer = Trainer(
    model= model,
    args=training_args,
    train_dataset = train_dataset,
    eval_dataset =  test_dataset,
    compute_metrics = compute_metrics,
)

trainer.train()
```

```python
eval_result = trainer.evaluate(eval_dataset=test_dataset)
print(eval_result)
```

```python
# 평가 데이터셋에 대한 예측 수행
predictions = trainer.predict(test_dataset)
predicted_labels = predictions.predictions.argmax(axis=-1)

# 예측 결과를 데이터프레임에 추가
test_texts = test_df['texts']  # 원본 텍스트 가져오기
results_df = pd.DataFrame({
    'text': test_texts,
    'true_label': test_df['labels'],
    'predicted_label': predicted_labels
})

print(results_df)
```
